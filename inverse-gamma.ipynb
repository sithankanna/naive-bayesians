```python
import numpy as np
import pandas as pd
from scipy.stats import invgamma
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (15,8)
plt.rcParams["font.size"] = 15
```

```python
x = np.linspace(start=0, stop=10, num=50)
x = np.round(x, 1)

n0 = 5
T0 = 60

for conf in range(2, 11):
    alpha = n0*conf/10
    beta = T0*conf/10
    df = pd.DataFrame(invgamma.pdf(x, a=alpha, scale=1/beta), index=x)
    plt.plot(df)
    plt.title(f"Confidence:{conf/10}, alpha={alpha}, beta={beta}, mean:{beta/(alpha)}")
#     plt.ylim()
    plt.show()
```

```python
df.to_clipboard()
```

```python
alpha = 3
beta = 2 
y = invgamma.rvs(a=alpha, scale=beta, size=500000)
df = pd.DataFrame(invgamma.pdf(x, a=alpha, scale=beta), index=x)

plt.plot(df)
plt.title(f"Theoretical Mean {beta/(alpha-1):.3f}, Emperical Mean {np.mean(y):.3f}")

```
